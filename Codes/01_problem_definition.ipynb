## 1.1 Problem Statement

Access to safe drinking water is one of the most critical global health challenges. According to the **World Health Organization (WHO)**, approximately **2 billion people** worldwide consume contaminated water, leading to diseases such as cholera, typhoid, and dysentery.

Traditional water quality testing is:
- **Expensive** — requires laboratory equipment and trained staff
- **Slow** — results can take days
- **Inaccessible** — especially in rural or developing regions

**Machine learning-based classification** offers a fast, scalable, and cost-effective alternative by predicting potability from physicochemical measurements.

### Goal
> Build a binary classification model that predicts whether a water sample is **potable (1)** or **non-potable (0)** based on its physicochemical properties.

## 1.2 Input Features (Physicochemical Properties)

| Feature | Description | WHO Safe Limit |
|---|---|---|
| **pH** | Measure of acidity/basicity (0–14) | 6.5 – 8.5 |
| **Hardness** | Calcium/Magnesium content (mg/L) | < 300 mg/L |
| **Solids (TDS)** | Total dissolved solids (ppm) | < 500 ppm |
| **Chloramines** | Disinfectant concentration (ppm) | < 4 ppm |
| **Sulfate** | Sulfate ion concentration (mg/L) | < 250 mg/L |
| **Conductivity** | Electrical conductivity (μS/cm) | < 400 μS/cm |
| **Organic Carbon** | Organic matter content (ppm) | < 2 ppm |
| **Trihalomethanes** | THM concentration (μg/L) | < 80 μg/L |
| **Turbidity** | Clarity of water (NTU) | < 5 NTU |

**Target:** `Potability` — 1 (Safe to drink), 0 (Not safe)

## 1.3 Literature Review

### Key Studies

**1. Bhatt & Bhatt (2020)**  
*"Machine Learning for Water Quality Prediction"*  
- Applied Random Forest and SVM on Indian groundwater samples
- RF achieved 87% accuracy; identified pH and TDS as top features
- Limitation: No class imbalance handling

**2. Hameed et al. (2022)**  
*"Comparative Analysis of ML Models for Drinking Water Quality"*  
- Compared Logistic Regression, Decision Tree, RF, XGBoost
- XGBoost outperformed with AUC of 0.91
- Used SMOTE for imbalanced classes

**3. Dawood et al. (2021)**  
*"Deep Learning vs. Classical ML for Water Potability"*  
- Found that well-tuned classical models (RF, XGB) match DL performance
- Classical models preferred for interpretability in regulatory contexts

### Research Gap Addressed in This Project
- Most studies do not combine **SMOTE + hyperparameter tuning + SHAP explainability**
- No interactive deployment for real-time prediction
- This project addresses all three gaps

## 1.5 Success Metrics

Given the **public health** context, **False Negatives are more dangerous** than False Positives:
- A FN = predicting unsafe water as safe → someone drinks contaminated water
- A FP = predicting safe water as unsafe → unnecessary caution (less harmful)

**Primary metric:** F1-Score (balances precision & recall)  
**Secondary metrics:** ROC-AUC, Recall (especially for class 1 = potable)  
**Target:** F1 ≥ 0.70, ROC-AUC ≥ 0.80
